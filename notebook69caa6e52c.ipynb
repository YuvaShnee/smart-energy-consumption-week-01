```python
#Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Step 2: Load datasets
train_path = "/kaggle/input/energy-consumption-dataset-linear-regression/train_energy_data.csv"
test_path  = "/kaggle/input/energy-consumption-dataset-linear-regression/test_energy_data.csv"

df_train = pd.read_csv(train_path)
df_test  = pd.read_csv(test_path)

print("Train shape:", df_train.shape)
print("Test shape :", df_test.shape)
```

```python
#Step 3: View first few rows
print("\n--- TRAIN HEAD ---")
print(df_train.head())

print("\n--- TEST HEAD ---")
print(df_test.head())
```

```python
#Step 4: Dataset info
print("\n--- TRAIN INFO ---")
print(df_train.info())

print("\n--- TEST INFO ---")
print(df_test.info())
```

```python
#Step 5: Summary statistics
print("\n--- TRAIN DESCRIBE ---")
print(df_train.describe(include="all").transpose())

print("\n--- TEST DESCRIBE ---")
print(df_test.describe(include="all").transpose())
```

```python
#Step 6: Check for missing values
print("\nMissing values in TRAIN:")
print(df_train.isnull().sum())

print("\nMissing values in TEST:")
print(df_test.isnull().sum())
```

```python
#Step 7: Define Features (X) and Target (y)

target_col = "Appliances"   # Change this if needed

if target_col in df_train.columns:
    X_train = df_train.drop(columns=[target_col])
    y_train = df_train[target_col]

    print("\n Features (X) shape:", X_train.shape)
    print(" Target (y) shape :", y_train.shape)
else:
    print(f"\n Target column '{target_col}' not found in TRAIN dataset. Check column names again.")
```
